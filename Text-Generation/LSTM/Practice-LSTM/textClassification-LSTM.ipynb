Importing the required libraries

```python
import pandas as pd
import numpy as np
import joblib
```

Importing the dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/LSTM/Practice-LSTM/emotion_detection_dataset.csv')
```

```python
dataset.head()
```

```python
dataset['Label'].value_counts()
```

Splitting input and output

```python
x = dataset['Text']
y = dataset['Label']
```

```python
x.head()
```

```python
y.head()
y.shape
```

Building the Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec = TfidfVectorizer()
x = vec.fit_transform(x)
x.toarray().shape
```

```python
print(x)
```

DataFraming x

```python
x = pd.DataFrame(x.toarray(), columns = vec.get_feature_names_out())
```

```python
x.shape
```

Encoding y

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(np.array([dataset['Label']]).reshape(-1, 1))
```

```python
encoded_data.toarray().shape
```

```python
print(encoded_data)
```

DataFramming y

```python
y = pd.DataFrame(encoded_data.toarray(), columns = encoder.get_feature_names_out())
```

```python
y.head()
```

Building the model

Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

Reshape the x in the formate of (No. of samples, time_stamp, features)

```python
x_train = x_train.values.reshape(36, 298, 1)
x_test = x_test.values.reshape(9, 298, 1)
x.head()
```

```python
model = Sequential()
```

```python
model.add(LSTM(1000, activation='relu', input_shape=(298, 1)))
model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=10)
```

Predicting the results

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
out_max = np.zeros_like(y_pred)
for i in range(y_pred.shape[0]):
  out = np.argmax(y_pred[i])
  out_max[i][out] = 1
```

```python
y_pred = encoder.inverse_transform(y_pred)
```

```python
print(y_pred)
```

Creating classification report

```python
from sklearn.metrics import classification_report
print(classification_report(encoder.inverse_transform(y_test), y_pred))
```

Exporting the model and

```python
model.save('emotionClassiLSTM.h5')
joblib.dump(vec, 'vec.pkl')
joblib.dump(encoder, 'encoder.pkl')
```

Building the local server

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Dec 20 21:40:15 2023

@author: Ramkumar
"""

from flask import Flask, request
import numpy as np
from tensorflow.keras.models import load_model
import joblib
import pandas as pd

model = load_model("C:/Users/Ramkumar/Downloads/emotionClassiLSTM.h5")
vec = joblib.load("C:/Users/Ramkumar/Downloads/vec.pkl")
encoder = joblib.load("C:/Users/Ramkumar/Downloads/encoder.pkl")

app = Flask(__name__)
@app.route('/', methods=['POST'])

def prediction():
    data = request.get_json(force=True)
    data = data['test']
    data = vec.transform(data)
    exc_data = data
    data = pd.DataFrame(data.toarray(), columns=vec.get_feature_names_out())
    data = data.values.reshape(1, data.shape[1], 1)
    pred = model.predict(data)

    out_max = np.zeros_like(pred)
    for i in range(pred.shape[0]):
        out = np.argmax(pred[i])
        out_max[i][out] = 1

    result = encoder.inverse_transform(out_max)

    return result[0][0]

app.run(host='0.0.0.0')
```
