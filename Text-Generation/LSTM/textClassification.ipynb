```python
x = dataset['text']
y = dataset['label']
```

Implementing Count Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

Droping the stoping words

```python
vec = TfidfVectorizer(stop_words = 'english')
newdata = vec.fit_transform(x)
newdata.toarray().shape
```

```python
newdata = pd.DataFrame(newdata.toarray(), columns=vec.get_feature_names_out())
```

```python
newdata
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
label_data = encoder.fit_transform(dataset[['label']])
```

```python
y = pd.DataFrame(label_data.toarray(), columns=encoder.get_feature_names_out())
y.head()
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(Dense(1000, activation='relu', input_shape=(29126,)))
model.add(Dense(1000, activation='relu'))
model.add(Dense(5, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(newdata, y, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model.fit(x_train, y_train, epochs=1) #epoch = 1000
```

```python
output = model.predict(x_test)
```

Making in the form of OneHotEncoded

```python
out_max = np.zeros_like(output)
for i in range(output.shape[0]):
  out = np.argmax(output[i])
  out_max[i][out]=1
```

```python
out_max
```

```python
out_max = encoder.inverse_transform(out_max)
```

```python
out_max
```

```python
y_test = encoder.inverse_transform(y_test)
```

Making the classification report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, out_max))
```

Exporting the model

```python
import joblib

model.save('tcud.h5')
joblib.dump(encoder, 'encoder.pkl')
joblib.dump(vec, 'vec.pkl')
```
