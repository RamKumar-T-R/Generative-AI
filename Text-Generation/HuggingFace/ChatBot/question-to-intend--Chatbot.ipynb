```python
import json
```

```python
data = open('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/Text-Generation/ChatBot/Qus_data.txt').read()
data
```

```python
data = data.replace("'", "\"")
print(data)
data = json.loads(data)
```

```python
x = []
y = []
```

```python
intent = list(data.keys())

for intent_name in intent:
  for sample in data[intent_name]:
    x.append(sample)
    y.append([intent_name])
```

```python
print(x)
print(y)
```

Tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words = 100)
tokenizer.fit_on_texts(x)
tokenizer.index_word
```

```python
tok_data = tokenizer.texts_to_sequences(x)
tok_data
```

Padding the sequences

```python
pad_data = pad_sequences(tok_data, maxlen=10)
pad_data
```

Encoding y

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
new_y = encoder.fit_transform(y)
new_y = new_y.toarray()
```

Building the model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(input_dim = 100, output_dim = 150, input_length = 10))
model.add(LSTM(50, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(pad_data, new_y, epochs = 1000)
```

Exporting the model and the pkl files

```python
model.save('ques_to_intend.h5')
```

```python
import joblib
joblib.dump(tokenizer, 'tokenizer.pkl')
joblib.dump(encoder, 'encoder.pkl')
```
