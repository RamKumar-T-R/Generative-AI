```python
import gensim
```

```python
model = gensim.models.keyedvectors.load_word2vec_format('wiki-news-300d-1M.vec')
```

```python
out = model['happy']
```

```python
model.most_similar(positive=['happy'], topn=5)
```

```python
model.most_similar(positive=['man'], topn = 5)
```
