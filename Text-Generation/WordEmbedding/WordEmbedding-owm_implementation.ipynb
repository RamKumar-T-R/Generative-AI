```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
text = ["hello python", "hey there", "hello world python code"]
```

Apply tokenizer

```python
token = Tokenizer(num_words=100)
token.fit_on_texts(text)
```

```python
seq = token.texts_to_sequences(text)
```

```python
seq
```

Padding the dataset

```python
pad_seq = pad_sequences(seq, padding='post')
```

```python
pad_seq
```

Constructing the embedding layer

```python
model = Sequential()
model.add(Embedding(input_dim = 10, output_dim = 150, input_length = 4))   # input_dim >= unique vocabularies present in dataset
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

```python
pad_seq[0, :]
```

```python
out = model.predict(pad_seq[0, :])
```
