Importing the required libraries

```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
import json
```

Loading the models and plk files

```python
from tensorflow.keras.models import load_model

model = load_model('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/Text-Generation/tcwelstm.h5')
```

```python
import joblib

encoder = joblib.load('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/Text-Generation/encoder.pkl')
token = joblib.load('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/Text-Generation/token.pkl')
file = open('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/Text-Generation/config.txt', 'r')
config = file.read()
```

```python
config = json.dumps(config)
```

```python
text_data = """2016 Rio Olympics bronze medallist Sakshi Malik has announced quitting wrestling following Sanjay Singh being elected as president of Wrestling Federation of India (WFI).

“Yeh ladai dil se ladi (We fought this battle with our heart). In the end, we slept for 40 days on the roads but I’d like to thank the several people of our country who came to support us during the protests earlier this year. If Brij Bhushan Singh’s business partner and a close aide is elected as the president of WFI, I quit wrestling…” Malik told members of the media in a press conference in New Delhi on Thursday before she walked out in tears with those present at the venue appealing her to reconsider her decision."""
```

```python
tok_seq = token.texts_to_sequences([text_data])
```

```python
tok_seq
```

```python
tok_pad = pad_sequences(tok_seq, maxlen=300)
```

```python
output = model.predict(tok_pad)
```

```python
output
```

```python
out_max = np.zeros_like(output)
for i in range(out_max.shape[0]):
  out = np.argmax(output[i])
  out_max[i][out] = 1
```

```python
out_max
```

```python
result = encoder.inverse_transform(out_max)
```
