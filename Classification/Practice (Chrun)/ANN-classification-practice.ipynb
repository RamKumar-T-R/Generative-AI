Bank Churn Prediction

Importing the required libraries

```python
import pandas as pd
import numpy as np
```

Importing the dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GEN_AI-Intern/Classification/Practice (Chrun)/Bank_churn_modelling.csv')
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['RowNumber', 'CustomerId', 'Surname'], axis='columns')
```

```python
dataset.head()
```

Data preporcessing - encoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

geo_encoder = OneHotEncoder()
gen_encoder = LabelEncoder()
exit_encoder = OneHotEncoder()
```

Transforming

```python
dataset['Gender'] = gen_encoder.fit_transform(dataset['Gender'])
```

```python
geo_data = geo_encoder.fit_transform(dataset[['Geography']])
geo_data = pd.DataFrame(geo_data.toarray(), columns = geo_encoder.get_feature_names_out())
dataset1 = pd.concat([dataset, geo_data], axis='columns')
```

```python
dataset1.head()
```

```python
exit_data = exit_encoder.fit_transform(dataset[['Exited']])
exit_data = pd.DataFrame(exit_data.toarray(), columns=exit_encoder.get_feature_names_out())
dataset2 = pd.concat([dataset1, exit_data], axis='columns')
```

```python
dataset2.head()
```

```python
dataset2 = dataset2.drop(['Geography', 'Exited'], axis='columns')
```

```python
dataset2.head()
```

Splitting the dataset into input and output

```python
x = dataset2.drop(['Exited_0', 'Exited_1'], axis='columns')
y = pd.concat([dataset2['Exited_0'], dataset2['Exited_1']], axis='columns')
```

Splitting the datset into train and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

Training the model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(30, activation='relu', input_shape=(12, )))
model.add(Dense(60, activation='relu'))
model.add(Dense(90, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
y_pred = model.predict(x_test)
```

```python
out_max = np.zeros_like(y_pred))
for i in range(y_pred.shape[0]):
  out = np.argmax(y_pred[i])
  out_max[i][out] = 1
```
